## Task 3 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from sklearn.datasets import load_linnerud
from sklearn.preprocessing import StandardScaler
```

```python
# Load the Linnerud dataset
data = load_linnerud()

# Create a DataFrame for easy data manipulation
df = pd.DataFrame(data.data, columns=data.feature_names)
```

## a

```python
# Compute basic statistics for the input features
input_stats = df.describe()

# Add the characteristics of the target features
input_stats["Chins Mean"] = data.target[:, 0].mean()
input_stats["Situps Mean"] = data.target[:, 1].mean()
input_stats["Jumps Mean"] = data.target[:, 2].mean()

input_stats["Chins Std"] = data.target[:, 0].std()
input_stats["Situps Std"] = data.target[:, 1].std()
input_stats["Jumps Std"] = data.target[:, 2].std()

# Print the characteristics of the features
print("Characteristics of the Linnerud dataset features:")
print(input_stats)
```

## b

```python
# Normalize the input features using StandardScaler
scaler = StandardScaler()
normalized_data = scaler.fit_transform(df)

# Create a DataFrame for the normalized data
normalized_df = pd.DataFrame(normalized_data, columns=data.feature_names)

# Suppress FutureWarnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Plot the probability density functions
plt.figure(figsize=(15, 5))

# Original Data
plt.subplot(1, 2, 1)
plt.title("Original Data")
for feature in data.feature_names:
    sns.kdeplot(df[feature], label=feature, fill=True)
plt.xlabel("Value")
plt.ylabel("Density")

# Normalized Data
plt.subplot(1, 2, 2)
plt.title("Normalized Data")
for feature in data.feature_names:
    sns.kdeplot(normalized_df[feature], label=feature, fill=True)
plt.xlabel("Value")
plt.ylabel("Density")

plt.tight_layout()
plt.legend()
plt.show()
```

### Summary:
By comparing these two figures, we can observe how the data distribution changes after normalization. The normalized data should have a mean of 0 and a standard deviation of 1 for all features. The KDE plots provide insights into the density of data points and how they are spread across different feature values.
This figures are useful for assessing the impact of normalization on our data and for understanding how features are distributed in the dataset.
